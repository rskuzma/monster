# EDA of Monster Jobs Data
- Look at columns to see which ones could be valuable in MVP

```python
import pandas as pd
```

```python
path = '/Users/richardkuzma/coding/analysis/monster/data/raw/'
filename = 'monster_jobs_sample.csv'
df = pd.read_csv(path+filename)

# create IDs to keep track of jobs, move titles and descriptions to the left for ease
df['id'] = df.index
df = df[['id', 'job_title', 'job_description'] + [c for c in df if c not in ['id', 'job_title', 'job_description']]]
df.head()
```

```python
df.info()
```

### Sector
- Sector may be useful to human-validate skills / matching, but some low-freq ones are pages long
- top 25 sectors by frequency account for 16,000/22,000 job postings

```python
sectors = df['sector'].value_counts()

sector_total = 0
num = 25
for i in range(0, num):
    sector_total += sectors[i]
print('Top {} sectors by frequency account for {} job records'.format(num, sector_total))

print('\n\n' + '-'*10 +' Top 50 sectors: ' + '-'*10 + '\n')
print(sectors.head(50))
```

## Locations
- Will need to be cleaned with regex and aggregated. Maybe just group by state

```python
locs = df['location'].value_counts()

location_total = 0
num = 50
for i in range(0, num):
    location_total += locs[i]
print('Top {} locations by frequency account for {} job records'.format(num, location_total))

print('\n\n' + '-'*10 +' Top 50 locations: ' + '-'*10 + '\n')
print(locs.head(50))
```
